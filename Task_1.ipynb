```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load Dataset
df = sns.load_dataset('iris')
# Check basic info
print(df.shape)
print(df.columns)
print(df.head())
# To understand Data
df.info()
df.describe()
# Scatter Plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
# Histogram
df.hist(figsize=(10,8))
plt.show()
# Box Plot
sns.boxplot(data=df)
plt.show()
```
